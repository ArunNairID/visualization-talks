## How to use Ggplot?

```python
from ggplot import *
```

## An example

```python
beef_plot = ggplot(aes(x='date', y='beef'), data=meat) +\
            geom_line() +\
            stat_smooth(colour='blue', span=0.2)
beef_plot
```

```python
#Â Different themes

beef_plot + theme_bw()
```

```python
diamond_plot = ggplot(diamonds, aes(x='carat', y='price', fill='cut')) +\
        geom_point(position = 'jitter')+\
        scale_color_brewer(type = 'seq', palette = 'Greens') +\
        stat_smooth(colour='black', span=0.4) + \
        xlab("Carats") + xlim(low=0, high=5.0) + \
        ylim(low=0, high=25000) +\
        scale_x_continuous(breaks=(0,3.0,5.0)) + \
        ylab("Price") +\
        scale_y_continuous(breaks=(0,10000,20000)) + \
        ggtitle("Diamonds") +\
        theme_gray() + \
        facet_wrap('cut')
diamond_plot
```
